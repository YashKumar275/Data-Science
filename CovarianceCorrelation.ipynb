# Covariance 

Covariance measures how two variables vary in tandem from their means.

```python
#hard way

import numpy as np
import matplotlib.pyplot as plt

def de_mean(x):
    xmean = np.mean(x)
    for i in x:
        return i - xmean
    
def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

plt.scatter(pageSpeeds, purchaseAmount)
plt.show()

cov = covariance(pageSpeeds, purchaseAmount)
print("Covariance:", cov)
```

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

plt.scatter(pageSpeeds, purchaseAmount)
plt.show()

cov = covariance(pageSpeeds, purchaseAmount)
print("Covariance:", cov)
```

```python
np.cov(pageSpeeds, purchaseAmount)
```

# Correlation
Correlation normalizes everything by their standard deviations, giving an easier to understand value that ranges from -1 (for a perfect inverse correlation) to 1 (for a perfect positive correlation).

```python
def correlation(x,y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y) / stddevx / stddevy

cor = correlation(pageSpeeds, purchaseAmount)
print("Correlation:", cor)
```

numpy can return a matrix of the correlation coefficients between every combination of the arrays passed in:

```python
np.corrcoef(pageSpeeds, purchaseAmount)
```
