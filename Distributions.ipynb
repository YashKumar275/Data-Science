# Uniform Distribution
A flat constant probability of a value occuring within a given range.

```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-10.0, 10.0, 100000)  #lower value= -10, higher value= 10, no. of values= 100000
plt.hist(values, 50)
plt.show()
```

# Probability Distribution Function
These are used on continuous data.

## Normal / Gaussian
Visualize the probability function

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-4, 4, 0.001)
plt.plot(x, norm.pdf(x))        #pdf = probabilty density function
```

Generate some random numbers with a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 5.0     #mean
sigma = 2.0  #standard deviation
values = np.random.normal(mu, sigma, 10000)    #10000 is specified number of data points
plt.hist(values, 50)    #50 buckets
plt.show()
```

## Exponential PDF / "Power Law"

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

# Probablity Mass Functions
These work on discrete data.

## Binomial Probablity Mass Function

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5      #Shape parameters (n: number of trials, p: probability of success)
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

## Poisson Probability Mass Function
It predicts the future outcome

Example: A website gets on an average of 500 visits per day. What's the odds of getting 550?

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```

# Percentile

```python
import numpy as np
import matplotlib.pyplot as plt

val = np.random.normal(0, 0.5, 1000)

plt.hist(val, 50)
plt.show()
```

```python
np.percentile(val, 50)  #50th percentile
```

```python
np.percentile(val, 90)  #90th percentile
```

```python
np.percentile(val, 99)  #99th percentile
```

```python
np.percentile(val, 25)  #25th percentile
```

# Moments
Measures the shape of a probability density function.

The first moment is the mean.
The second moment is the variance.
The third moment is the skew.
The fourth moment is the kurtosis.

## First Moment (Mean) 

```python
np.mean(val)
```

## Second Moment (Variance)

```python
np.var(val)
```

## Third Moment (Skew)

```python
import scipy.stats as sci

sci.skew(val)
```

## Fourth Moment (Kurtosis)
Describes the shape of the tail

```python
sci.kurtosis(val)
```
