# Linear Regression
Fit a line to a data set of observations.

It uses least squares method. It minimizes the squared-error between each point and the line 

```python
!pip install numpy
!pip install matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt

websiteSpeed = np.random.normal(5.0, 2.5, 1000)
purchaseAmount = 100 - (websiteSpeed + np.random.normal(0, 0.2, 1000))*3

plt.scatter(websiteSpeed, purchaseAmount)
```

Least square minimizes the sum of squared errors.

```python
!pip install scipy
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(websiteSpeed, purchaseAmount)
```

### R-Squared 
Aka Coefficient of Determination

It measures, the fraction of the total variation in Y that is captured by the model.

It measures how well our line fits best to out data.

R-square = 1.0 - (sum of squared errors / sum of squares variation from mean)

Ranges from 0 to 1.

0 is bad (none of the variance is captured), 1 is good (all  of the variance is captured)

R-squared value shows a really good fit:

```python
result = r_value ** 2
print(result)
```

Let's use the slope and the intercept we got from the regression to plot predicted values vs. observed:

```python
def predict(x):
    return slope * x + intercept

bestfitline = predict(websiteSpeed)

plt.scatter(websiteSpeed, purchaseAmount)
plt.plot(websiteSpeed, bestfitline, c='r')   #here c means color, which is assigned as red
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

websiteSpeed = np.random.normal(100.0, 2, 1000)
purchaseAmount = 100 - (websiteSpeed + np.random.normal(100, 5, 1000))*3

plt.scatter(websiteSpeed, purchaseAmount)
```

# Polynomial Regression

Not all relations are linear.

```python
np.random.seed(2)     #seed(2) will help in keep the result constant of the random
pageSpeed = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeed

plt.scatter(pageSpeed, purchaseAmount)
```

numpy has a handy polyfit function, which can construct an nth-degree polynomial model of the data that minimizes squares error.

```python
x = np.array(pageSpeed)
y = np.array(purchaseAmount)

p4 = np.poly1d(np.polyfit(x, y, 4))   #poly1d takes the array of coefficients and returns a polynomial object
```

We'll visualize our original scatter plot, together with a plot of our predicted values using the polynomial for page speed times ranging from 0-7 seconds:

```python
xp = np.linspace(0, 7, 1000)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

R-squared error:

```python
!pip install --upgrade pip
```

```python
!pip install scikit-learn
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p4(x))

print(r2)
```

```python
np.random.seed(3)     #seed(3) will help in keep the result constant of the random
pageSpeed = np.random.normal(5.0, 2.0, 1000)
purchaseAmount = np.random.normal(50.0, 20.0, 1000) / pageSpeed

plt.scatter(pageSpeed, purchaseAmount)
```

```python
x = np.array(pageSpeed)
y = np.array(purchaseAmount)

p8 = np.poly1d(np.polyfit(x, y, 8))   #poly1d takes the array of coefficients and returns a polynomial object
```

```python
xp = np.linspace(0, 7, 1000)
plt.scatter(x, y)
plt.plot(xp, p8(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p8(x))

print(r2)
```

```python
import pandas as pd

df = pd.read_excel('http://cdn.sundog-soft.com/Udemy/DataScience/cars.xls')
```

```python
%matplotlib inline
df1 = df[['Mileage', 'Price']]
bins = np.arange(0, 50000, 10000)       #divide into 10000 chunks of data ranging from 0 to 50000
groups = df1.groupby(pd.cut(df1['Mileage'], bins)).mean()
print(groups.head())
groups['Price'].plot.line()
```

```python
!pip install statsmodels
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

x = df[['Mileage', 'Cylinder', 'Doors']]
y = df['Price']

x[['Mileage', 'Cylinder', 'Doors']] = scale.fit_transform(x[['Mileage', 'Cylinder', 'Doors']].values)

#Add a constant column to our model so we can have a Y-intercept
x = sm.add_constant(x)

print(x)

est = sm.OLS(y, x).fit()

print(est.summary())
```

The table of coefficients above gives us the values to plug into an equation of form: B0 + B1 * Mileage + B2 * Cylinders + B3 * Doors

```python
y.groupby(df.Doors).mean()
```

```python
scaled = scale.transform([[45000, 8, 4]])
scaled = np.insert(scaled[0], 0, 1)    #Need to add that constant column in again.
print(scaled)
predicted = est.predict(scaled)
print(predicted)
```
